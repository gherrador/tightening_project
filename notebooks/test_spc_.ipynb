#### I-Chart (Individuals) del mes para STEPs con alertas

Qué buscamos encontrar con este grafico:

- Torque vs tiempo
- Límites UCL/LCL
- Puntos alertados (MR o I)

Para ello revisaremos datos solo para steps cuando:

- Solo para STEPs con n_alerts > 0
- Prioriza top-N por n_alerts

```python
def plot_spc_i_chart(df_points, df_limits, step_id):
    d = df_points[df_points["STEP_ID"] == step_id].sort_values("DateTime")
    l = df_limits[df_limits["STEP_ID"] == step_id].iloc[0]

    plt.figure(figsize=(12,4))
    plt.plot(d["DateTime"], d["FinalTorque"], lw=1)
    plt.axhline(l["UCL"], ls="--")
    plt.axhline(l["LCL"], ls="--")
    plt.axhline(l["Xbar"], ls=":")
    plt.scatter(
        d.loc[d["is_alert"]==1,"DateTime"],
        d.loc[d["is_alert"]==1,"FinalTorque"],
        c="red", s=15, zorder=3
    )
    plt.title(f"SPC I-Chart – STEP {step_id}")
    plt.show()
```
